````python
y
````

# Vamos falar sobre reiniciar o Kernel e as variáveis.


Sempre bom durante a utilização de um Notebook salvar Checkpoints, para caso haja algum problema com o computador, ou a luz caia, ter esse checkpoint salvo e não perder muita informação feita.

### Para salvar Checkpoints:
```
Ctrl + S
```

Quando saímos de algum notebook, e vamos olhar pelo jupyter em si, podemos ver os arquivos de notebook que ainda estão rodando o python mesmo que fechados. Aqueles que estão rodando, estão com o simbolo do notebook em **verde**. Enquanto o que estão parados estão em **preto**.

Só fechando a guia, não implica em fechar o notebook, caderno, só fechará de fato, quando na guia Jupyter vc fizer o _Shutdown_ do caderno, ou na própria guia ir na guia _File_ e ir em _Close and Halt_.

Outra forma de reiniciar o arquivo é indo na guia Kernel e colocando para _restart_ que vai funcionar como fechar e abrir novamente, e toda nova linha de código que será rodada será a partir da linha 1. Ele irá deletar todas variáveis criadas anteriormente e vai partir do zero.

> Uma sugestão que é dada no vídeo é tentar escrever e rodar suas células de código em ordem  correta, na vertical, para que quando for rodar de uma vez, não haja nenhum erro, por causa de ordem.

Ele sugere rodar uma de cada vez, mas com notebook muito grande é melhor utilizar a configuração da guia _Kernel_ de _Restart and Run All_. 

> Sempre tente colocar as linhas de código em ordem crescente de cima para baixo na página, para que nã haja erro de variáveis e funções não definidas.

# Interrompendo o Kernel

Quando fazendo um incidente que leve ao programa a rodar infinitamente (como mostraremos no exemplo abaixo),  temos como parar o programa interrompendo o kernel, basta apertar o quadrado preto ao lado do run ou fazer o shortcut de aperta o I+I, na guia kernel é possível ver.

````python
import time 
while True:
    print(1)
    time.sleep(1)
````

Aqui é mostrado uma situação em que selecionado a célula, interrompemos o kernel pois se não iria ficar para sempre funcionando. (é utilizado o tiime.sleep para não criar muita informação desnecessária).

> Um resumo dos shortcuts tbm é possível ser encontrado ao lado da definição do tipo de célula, embaixo a direita da guia Help tem um mini keyboard, que se clicar nele mostra todos os possíveis shortcuts.

Ao mesmo tempo em que falo isso, existe uma forma de ver todos os possíveis shortcuts dos dois modos, que é indo na guia _Help_ e depois indo para _Keyboard Shortcuts_. Ou basta apertar H no command mode. Ele mostrará todos os possíveis shortcuts, que não precisam saber de todos, mas com alguns já facilitará bastante a vida.
